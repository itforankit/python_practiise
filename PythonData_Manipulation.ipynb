```python
data6_6 = ['india 1998', 'big country', np.nan]
series_6_6 = pd.Series(data6_6)

def reverse_lower_case_words(text):
    if pd.isna(text):
        return text
    words = text.split()
    reversed_words = [word[::-1] if word.islower() else word for word in words]
    return ' '.join(reversed_words)

reversed_series = series_6_6.apply(reverse_lower_case_words)

print(reversed_series)
```
