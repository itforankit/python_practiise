```python
from pytube import YouTube

# Replace this URL with the link to the Django tutorial video on YouTube
video_url = "https://www.youtube.com/playlist?list=PLxxA5z-8B2xk4szCgFmgonNcCboyNneMD"

try:
    yt = YouTube(video_url)
    video = yt.streams.get_highest_resolution()
    # Change the directory path where you want to save the video
    video.download("D:\\downloadvideos\\")
    print("Video downloaded successfully!")
except Exception as e:
    print("Error:", str(e))
```

```python
from pytube import Playlist

# Replace this URL with the link to the Django tutorial playlist on YouTube
playlist_url = "https://www.youtube.com/playlist?list=PLxxA5z-8B2xk4szCgFmgonNcCboyNneMD"

try:
    playlist = Playlist(playlist_url)
    for video in playlist.videos:
        video.streams.get_highest_resolution().download("D:\\downloadvideos\\")
        print(f"Downloaded: {video.title}")
    print("All videos downloaded successfully!")
except Exception as e:
    print("Error:", str(e))
```
