```python
import folium
import pandas as pd

data=pd.read_csv("Police_Department_Incidents_Year_2016_.csv")

#Display first 100 records from the data and play with parameters - tiles, 
#zoom_start etc. Map will be saved in BasicWebMap.html  view it in browser

data_100 = data.head(100)
```

```python
data.head(2)
```

```python
# Convert Date column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Sort the dataframe based on the Date column in descending order
sorted_data = data.sort_values(by='Date', ascending=False)

# Extract the top 7 unique dates
top_7_dates = sorted_data['Date'].unique()[:7]

top_7_dates
```

```python
# Create a map centered on San Francisco
map_sf = folium.Map(location=[37.7749, -122.4194],  zoom_start=15)
```

```python
for index, row in data_100.iterrows():
    folium.Marker([row['Y'], row['X']], popup=row['Category']).add_to(map_sf)
```

```python
# Save the map as BasicWebMap.html
map_sf.save("BasicWebMap.html")
```

```python
#For latest 7 days of data create WebMap of Crimes which are categorized as ROBBERY
from datetime import datetime, timedelta

# Convert Date column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Filter data for the latest 7 days and crimes categorized as "ROBBERY"
max_date_str = '2016-12-31 23:59:59'
max_date = datetime.strptime(max_date_str, '%Y-%m-%d %H:%M:%S')

seven_days_ago = max_date - timedelta(days=7)

filtered_data = data[(data['Date'] >= seven_days_ago) & (data['Category'] == 'ROBBERY')]

# Create a map centered on San Francisco
map_sf = folium.Map(location=[37.7749, -122.4194], zoom_start=12)

# Add markers for each "ROBBERY" incident
for index, row in filtered_data.iterrows():
    folium.Marker([row['Y'], row['X']], popup=row['Descript']).add_to(map_sf)

# Save the map as a HTML file
map_sf.save("Robbery_Crimes_Map.html")
```

```python
# Filter data for the latest 15 days and crimes categorized as "FRAUD" and "GAMBLING"
max_date_str = '2016-12-31 23:59:59'
max_date = datetime.strptime(max_date_str, '%Y-%m-%d %H:%M:%S')
fifteen_days_ago = max_date - timedelta(days=15)

filtered_data = data[((data['Category'] == 'FRAUD') | (data['Category'] == 'GAMBLING')) & (data['Date'] >= fifteen_days_ago)]

# Path to the Font Awesome solid SVG icon file
icon_path = 'iconfontawesome.svg'

# Create a map centered on San Francisco
map_sf = folium.Map(location=[37.7749, -122.4194], zoom_start=12)

# Add markers for each "FRAUD" and "GAMBLING" incident using Font Awesome icons
for index, row in filtered_data.iterrows():
    icon = folium.CustomIcon(icon_image='creativity.svg', icon_size=(30,30))
    folium.Marker([row['Y'], row['X']], icon=icon, popup=row['Descript']).add_to(map_sf)

# Save the map as a HTML file
map_sf.save("Fraud_Gambling_Crimes_Map.html")
```

```python
#BONUS ASSIGNMENT --  Display heatmap for Divvy Bikes.  Divvy Bikes runs 
#bike rental service in Chicago and their bike station geolocation data is shared. #
#Refer to BikesHeatMap.py and Divvy_Stations_2016_Q3.csv, 
#Divvy_Stations_2016_Q4.csv
from folium.plugins import HeatMap

# Load the Divvy bike station data for Q3 and Q4 2016
divvy_q3 = pd.read_csv("Divvy_Stations_2016_Q3.csv")
divvy_q4 = pd.read_csv("Divvy_Stations_2016_Q4.csv")

# Concatenate the Q3 and Q4 dataframes
divvy_data = pd.concat([divvy_q3, divvy_q4])

# Create a Folium map centered on Chicago
map_chicago = folium.Map(location=[41.8781, -87.6298], zoom_start=11)

# Convert latitude and longitude columns to a list of lists
locations = divvy_data[['latitude', 'longitude']].values.tolist()

# Create a heatmap layer using the locations data
heat_map = folium.plugins.HeatMap(locations, radius=15)

# Add the heatmap layer to the map
map_chicago.add_child(heat_map)

# Save the map as an HTML file
map_chicago.save("Divvy_Bikes_Heatmap.html")
```
