```python
import requests
r=requests.get("https://www.thomascook.in/holidays/india-tour-packages/kerala-tour-packages")
c=r.content
print(c)
```

```python
from bs4 import BeautifulSoup
soup =BeautifulSoup(c, "html.parser")
print(soup.prettify())
```

```python
links =[]
for link in soup.find_all('a'):
    #print(link)
    links.append(link.get('href'))
links
```

```python
for price in soup.find_all('span',class_='INR'):
    print(price)
    
```

```python
for itemprice in soup.find_all('span',itemprop_='price'):
    print(itemprice)
    
```

```python
from bokeh.plotting import output_file,show,figure
import pandas

dataset=pandas.DataFrame(columns=['X','Y'])
dataset['X']=[1,2,3,4,5]
dataset['Y']=[1,4,9,16,25]

p=figure(width=500, height=400)
p.circle(dataset['X'],dataset['Y'],size=5)
output_file("scatter_charts.html")
show(p)
```

```python
import cv2
# 1 for color and 0 for black and white
img=cv2.imread('penguins.jpg',1)
resized_image = cv2.resize(img, (int(img.shape[1]/3),int(img.shape[0]/3)))
cv2.imwrite("penguin_resize_1.jpg",resized_image)
```

```python
cv2.imshow("penguins",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
