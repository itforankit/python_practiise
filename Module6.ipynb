```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[6,4,15,6])
plt.show()
```

```python
import numpy as np
dic={'a':25,'b':70,'c':55,'d':90}
for i, key in enumerate(dic):
    plt.bar(i,dic[key])
plt.xlabel("Value(x)")
plt.ylabel("Fucntions [x,x2,x3]")
plt.title("Poly Graph")
plt.legend()
plt.xticks(np.arange(len(dic)),dic.keys())
```

```python
x=[5,4,5,6]
plt.plot(x, x, label='linear')
plt.plot(x, x*x, label='square')
plt.plot(x, x*x*x, label='cube')
plt.show()
```

```python
import folium
map=folium.Map(location=[30.309219,78.009529],zoom_start=10,tiles='openstreetmap')
schools=folium.FeatureGroup(name='schools')
for coordinate in ([30.309219,78.009529],[30.109219,78.109529],[30.308220,78.008519]):
    schools.add_child(folium.Marker(location=coordinate,popup="schools..",icon=folium.Icon(color="red",icon='cloud')))
map.add_child(schools)
map.save("folium.html")
map
```
